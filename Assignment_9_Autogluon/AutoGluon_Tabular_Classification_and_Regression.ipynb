```python
import pandas as pd

# Loading the Titanic dataset
titanic_data = pd.read_csv('Titanic-Dataset.csv')  # Update the file name if different
print(titanic_data.head())

# Loading the California Housing Prices dataset
housing_data = pd.read_csv('housing.csv')  # Update the file name if different
print(housing_data.head())
```

```python
titanic_data['FamilySize'] = titanic_data['SibSp'] + titanic_data['Parch'] + 1
titanic_data = pd.get_dummies(titanic_data, columns=['Sex', 'Embarked'])
```

```python
housing_data['total_bedrooms'].fillna(housing_data['total_bedrooms'].median(), inplace=True)
```

```python
from autogluon.tabular import TabularPredictor

# Titanic survival prediction
titanic_predictor = TabularPredictor(label='Survived').fit(titanic_data)

# California housing price prediction
housing_predictor = TabularPredictor(label='median_house_value').fit(housing_data)
```

```python
from autogluon.tabular import TabularPredictor
from sklearn.model_selection import train_test_split

# Assuming titanic_data and housing_data are already loaded and preprocessed

# Split the data into train and test sets
train_titanic, test_titanic = train_test_split(titanic_data, test_size=0.2, random_state=42)
train_housing, test_housing = train_test_split(housing_data, test_size=0.2, random_state=42)

# Train the models
titanic_predictor = TabularPredictor(label='Survived').fit(train_titanic)
housing_predictor = TabularPredictor(label='median_house_value').fit(train_housing)

# Evaluate the models on the test data
titanic_results = titanic_predictor.evaluate(test_titanic)
housing_results = housing_predictor.evaluate(test_housing)

print("Titanic model evaluation results:", titanic_results)
print("Housing model evaluation results:", housing_results)


```
