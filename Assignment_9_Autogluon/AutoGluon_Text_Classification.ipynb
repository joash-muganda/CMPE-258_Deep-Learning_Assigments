```python
import pandas as pd

# Assuming the dataset CSV is named 'text_classification_data.csv'
data = pd.read_csv('df_file.csv')
print(data.head())
```

```python
# Check the balance of the labels
label_counts = data['Label'].value_counts()
print(label_counts)

# Check for any missing values
missing_data = data.isnull().sum()
print("Missing values in each column:\n", missing_data)
```

```python
# Basic text cleaning
data['Text'] = data['Text'].str.lower().str.replace(r'[^\w\s]+', '')

# If any text entries are missing, consider dropping or filling them
data.dropna(subset=['Text'], inplace=True)
```

```python
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)
```

```python
!pip install autogluon.text
```

```python
from autogluon.text import TextPredictor

predictor = TextPredictor(label='Label', eval_metric='acc').fit(train_data=train_data)
```

```python
performance = predictor.evaluate(test_data)
print("Test Accuracy:", performance['acc'])
```

```python
predictions = predictor.predict(test_data)
print(predictions.head())
```
